### Step 2: Data Preperation

* Dropping irrelevant columns
* Identifying duplicate columns
* Renaming Columns
* Feature Creation
* Feature Selection

```python
# first way to filter the feature or create a subset of dataset
df = df[['coaster_name', 
       'Location',
       'Status',
       'Manufacturer', 
       'year_introduced', 
       'latitude', 
       'longitude', 'Type_Main',
       'opening_date_clean', 
       'speed_mph', 
        'height_ft',
       'Inversions_clean', 
       'Gforce_clean']].copy()

# second is by dropping the columns or feature
#  df.drop(['<Column_name>'], axis=1) # 1 for columns & 2 for row
```

```python
df.head()
```

```python
df.info()
```

```python
# changing the dtype
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
df.info()
```

```python
# Renaming the columns
df = df.rename(columns={'coaster_name':'Coaster_Name', 
                    'year_introduced':'Year_Introduced',
                    'opening_date_clean':'Opening_Date', 
                    'speed_mph':'Speed_mph',
                    'height_ft':'Height_ft', 
                    'Inversions_clean':'Inversions', 
                    'Gforce_clean':'Gforce'})
```

```python
# Treating the missing or Null values
df.isna().sum()
```

```python
# Treating the duplicate values
df.loc[df.duplicated()]
```

```python
# checking duplicate on subset of column
df.loc[df.duplicated(subset=['Coaster_Name'])].head()
```

```python
# checking an example of duplicate
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])] \
    .reset_index(drop=True).copy()
df.shape
```

### Step 3: Feature Understanding

(Univariate Analysis)

* Plotting Feature Distribution
> * Histogram
> * KDE
> * Boxplot

```python
df.Year_Introduced.value_counts()
```

```python
# Bar Plot
ax = df['Year_Introduced'].value_counts() \
        .head(10) \
        .plot(kind='bar', title='Top 10 Years Roller Coaster Introduced', color='darkred', label=True)

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
# Histogram
ax = df['Speed_mph'].plot(kind='hist',
                          bins=20, 
                          title='Coaster speed (mph)', color= 'g')

ax.set_xlabel('Speed (mph)')
```

```python
# KDE - Kernel Density Plot
ax = df['Speed_mph'].plot(kind='kde', 
                          title='Coaster speed (mph)', color= 'r', style='*')

ax.set_xlabel('Speed (mph)')
```

```python
# Boxplot
ax = df['Speed_mph'].plot(kind='box', title='Coasters speed distribution (mph)', label="")

ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Frequency')
```
